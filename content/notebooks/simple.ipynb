Lets download all level 5 tiles from the Map Quest Open Aerial project (http://developer.mapquest.com/web/products/open/map). These files are a part of the gist repository found at https://gist.github.com/6139282 (git clone from https://gist.github.com/6139282.git).

````python
import urllib
import itertools
import os

url_template = 'http://oatile1.mqcdn.com/tiles/1.0.0/sat/{z}/{x}/{y}.jpg'
tile_fname_template = "tiles/tile_{x}_{y}_{z}.jpg"
z = 5

for x, y in itertools.product(xrange(2 ** z), xrange(2 ** z)):
    fname = tile_fname_template.format(x=x, y=y, z=z)
    if not os.path.exists(fname):
        urllib.urlretrieve(url_template.format(x=x, y=y, z=z),
                            fname)
````

Now that we know how to load an individual image, we can create a biggus.ArrayAdapter which can defer any reading from disk until the array is actually requested. First, we need a class which can load our data when it is passed acceptable indexes via the __getitem__ method.

````python
class TileLoader(object):
    def __init__(self, img_fname):
        self.concrete_array = None
        self.img_fname = img_fname
        
        # The required attributes of a "concrete" array, acording to biggus,
        # are 'ndim', 'dtype', and 'shape'.
        self.ndim = 3
        self.dtype = np.uint8
        self.shape = (256, 256, 3)
    
    def __getitem__(self, slices): 
        # Implement the getitem/slicing interface which is the interface needed for biggus to create numpy arrays.
        if self.concrete_array is None:
            print 'loading image from {}'.format(self.img_fname)
            self.concrete_array = np.array(PIL.Image.open(self.img_fname))
        return self.concrete_array.__getitem__(slices)
    
    def unload(self):
        # Unloads the data, so that we can do some testing. This is not a necessary interface.
        self.concrete_array = None
````

A TileLoader instance can now be created by passing just the filename of the tile which is of interest. Notice how the data isn't actually loaded until we ask for it with the index notation.

````python
fname = tile_fname_template.format(x=15, y=10, z=5)
defered_img = TileLoader(fname)
````

We can now go ahead and create a biggus.ArrayAdapter instance with a TileLoader instance.

````python
import biggus
img = biggus.ArrayAdapter(TileLoader(fname))

print img
sub_img = img[::50, 30:35, 0]
````

Notice how the defered loading of the image array has persisted and is only loaded once the actual array is requested via the ```Array.ndarray()``` method.

````python
print sub_img.ndarray()
````

So we can now arbitrarily index an array which isn't loaded until we request the actual data. Whilst ```Array.ndarray()``` on sub_img in this case was achieved by first loading the entire image into memory and then subsequently indexing, biggus provides enough context so that it is possible for the cases where it is feasible to only read a subset of data from disk. One good example of this is with NetCDF variables.

Now that we can create ```biggus.Array``` instances (```biggus.ArrayAdapter``` is a subclass of ```biggus.Array```) we can use some of its more interesting features.


Lets create a ```biggus.LinearMosaic``` to represent a column of map tiles.

````python
def tile_adapter(x, y, z):
    fname = tile_fname_template.format(x=x, y=y, z=z)
    defered_img = TileLoader(fname)
    return biggus.ArrayAdapter(defered_img)

arrays = np.array([tile_adapter(x=15, y=y, z=5) for y in xrange(2 ** z)])
col = biggus.LinearMosaic(arrays,
                          axis=0)
````

````python
print col.shape
````

We can go ahead and slice a subset of the big column of images and visualise the resulting ndarray. Notice how only the necessary images from the 32 image mosaic are actually loaded.

````python
smaller_column = col[2000:3500]
plt.figure(figsize=(3, 10))
plt.imshow(smaller_column.ndarray())
````

The ```biggus.LinearMosaic``` will take any array like object in its mosaic, including other ```biggus.LinearMosaic``` instances, so we could now go ahead an make a horizontal mosaic of image columns to represent the whole dataset.

````python
def tile_array(z):
    columns = []
    for x in xrange(2 ** z):
        arrays = np.array([tile_adapter(x=x, y=y, z=z) for y in xrange(2 ** z)])
        columns.append(biggus.LinearMosaic(arrays,
                                           axis=0))
    return biggus.LinearMosaic(columns, axis=1)

z5 = tile_array(z=5)
````

````python
print z5.shape
````

Whilst this array is not bigger than could actually fit in memory, it is worth remembering that we can access any subsection of this array (which is split into 32 * 32 separate files) easily without needing to allocate the full amount of memory (which in this case is ~192MB).

````python
def sizeof_fmt(num):
    for x in ['bytes','KB','MB','GB']:
        if num < 1024.0:
            return "%3.1f%s" % (num, x)
        num /= 1024.0
    return "%3.1f%s" % (num, 'TB')

def sizeof_array(a):
    return sizeof_fmt(a.dtype().itemsize * np.prod(a.shape))

print sizeof_array(z5)
````

Again, we can easily visualise a subset of this large array by using the numpy indexing notation.

````python
smaller_img = z5[2250:3000, 3700:4550]
plt.imshow(smaller_img.ndarray())
````

Lets update TileLoader so that the image is automatically retrieved from the Map Quest server on demand. The full code now necessary to generate a arbitrary z-level mosaic of map tiles is:

````python
import io
import urllib

import biggus


class TileLoader(object):
    """Re-implements the TileLoader, only this time downloads the tiles from the server on demand."""
    def __init__(self, x, y, z):
        self.concrete_array = None
        self.xyz = x, y, z
        
        # The required attributes of a "concrete" array, acording to biggus,
        # are 'ndim', 'dtype', and 'shape'.
        self.ndim = 3
        self.dtype = np.uint8
        self.shape = (256, 256, 3)
    
    def __getitem__(self, slices): 
        # Implement the getitem/slicing interface which is the interface needed for biggus to create numpy arrays.
        if self.concrete_array is None:
            print 'Downloading tile x={} y={} z={}'.format(*self.xyz)
            url = url_template.format(x=self.xyz[0], y=self.xyz[1], z=self.xyz[2])
            tile_fh = urllib.urlopen(url)
            image_file = io.BytesIO(tile_fh.read())
            self.concrete_array = np.array(PIL.Image.open(image_file))
        return self.concrete_array.__getitem__(slices)


def tile_array(z):
    columns = []
    for x in xrange(2 ** z):
        print 'Done: x=', x
        arrays = np.array([biggus.ArrayAdapter(TileLoader(x=x, y=y, z=z)) for y in xrange(2 ** z)])
        columns.append(biggus.LinearMosaic(arrays,
                                           axis=0))
    return biggus.LinearMosaic(columns, axis=1)
````
